## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
client_df.loc[:,'line_subtotal'] = client_df['unit_price'] * client_df['qty']

# Display the updated DataFrame
client_df.head()
```

```python
sys.path.append('shipping_calculator.py')
from shipping_calculator import calculate_shipping_price

# Create a column for shipping price.
# Assume a shipping price of $7 per pound for items over 50 pounds and $10 per pound for items 50 pounds or under.

client_df['shipping_price']  = client_df.apply(
    lambda row: calculate_shipping_price(row['unit_weight'], row['qty']), axis=1
)

client_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# Calculate the total price
client_df['total_price'] = client_df.apply(
    lambda x: (x['line_subtotal'] + x['shipping_price']) *1.0925, axis=1).round(2)

# Display the updated data
client_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

client_df['line_cost'] = (client_df['unit_cost'] * client_df['qty']) + client_df['shipping_price']

client_df.head()

```

```python
# Create a column for the profit of each line using line cost and line price

client_df['line_profit'] = client_df['total_price'] - client_df['line_cost']

client_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Function to sum the total price based on client_id
def client_totals(client_id, column):
    df = client_df.loc[client_df['client_id']== client_id , column]
    return round(df.sum(),2)

# Call function and display results
for client_id in top_client_ids:
    print(f"{client_id}: ${client_totals(client_id, 'total_price')}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.

top_clients_df = client_df[client_df['client_id'].isin(top_client_ids)]

# Group by 'client_id' and sum
summary = top_clients_df.groupby('client_id').agg(
    Client_Name=('client_name', 'first'),
    Total_Units_Purchased=('qty', 'sum'),
    Total_Shipping_Price=('shipping_price', 'sum'),
    Total_Revenue=('total_price', 'sum'),
    Total_Profit=('line_profit', 'sum')
).reset_index()

# Sort the summary DataFrame by Total_Revenue in descending order
summary = summary.sort_values(by='Total_Revenue', ascending=False)

# Display the sorted summary DataFrame
print("Summary DataFrame sorted by Total Revenue (descending):")
summary
```

```python
# Format the data and rename the columns to names suitable for presentation.
columns_to_convert = ['Total_Shipping_Price', 'Total_Revenue', 'Total_Profit']
summary[columns_to_convert] = summary[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Handle NaN values
summary[columns_to_convert] = summary[columns_to_convert].fillna(0)
# Define the money columns. 
money_columns =['Total_Shipping_Price', 'Total_Revenue', 'Total_Profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    # return f"${x / 1_000_000:.2f}" if pd.notnull(x) else "$0.00"
    return x / 1000000
# Apply the currency_format_millions function to only the money columns. 
summary[money_columns] = summary[money_columns].map(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
summary.rename(columns={
    'client_id': 'Client ID',
    'Total_Shipping_Price': 'Total Shipping Price (in millions)',
    'Total_Revenue': 'Total Revenue (in millions)',
    'Total_Profit': 'Total Profit (in millions)'
               }, inplace=True)

print(summary)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
# Sort the summary DataFrame by Total_Revenue in descending order
summary_by_total_revenue = summary.sort_values(by='Total Revenue (in millions)', ascending=False)

# Display the sorted summary DataFrame
print("Summary DataFrame sorted by Total Revenue (descending):")
summary_by_total_revenue
```
